```python
import pandas as pd
```

```python
response = pd.read_csv('marshall-islands-taiwan.csv')
marshall = pd.DataFrame(response)
marshall
```

```python
marshall
```

```python
marshall = marshall.drop([10])
marshall
```

```python
marshall = marshall.drop(['Notes'], axis=1)
```

```python
marshall['grants'] = (marshall['ROC Capital Grants (Projects?)']
                                              .str.replace("$","")
                                              .str.replace("million","")
)
marshall
```

```python
marshall['other'] = (marshall['Grants (Tax-Free?)']
                    .str.replace("$","")
                    .str.replace("million","")
                     .str.replace("NaN",'0')
                    )
marshall
```

```python
marshall = marshall.fillna(0)
```

```python
marshall.drop(['ROC Capital Grants (Projects?)'], axis=1)
marshall.drop(['Grants (Tax-Free?)'], axis=1, inplace=True)
marshall.drop(['Total'], axis=1, inplace=True)
marshall
```

```python
marshall['scholarship'] = (marshall['Scholarship']
                           .str.replace("$","")
                           .str.replace("million","")
)
marshall = marshall.drop(['Scholarship'], axis=1)
marshall = marshall.fillna(0)
marshall
```

```python
marshall['total'] = marshall.grants.astype(float) + marshall.other.astype(float) + marshall.scholarship.astype(float)
marshall.drop(['ROC Capital Grants (Projects?)'], axis=1, inplace=True)
marshall
```

```python
marshall.to_csv('marshall_clean.csv')
```

```python
nauru = pd.read_csv('nauru-taiwan.csv')
nauru
```

```python
nauru = nauru.fillna(0)
```

```python
nauru = nauru.drop([10])
nauru
```

```python
nauru['grants'] = (nauru['Grants (USD)']
                   .str.replace("$","")
                   .str.replace("million","")
                  )
nauru = nauru.drop(['Grants (USD)'], axis=1)
nauru
```

```python
nauru['grants-in-kind'] = (nauru['Grants In Kind (USD)']
                   .str.replace("$","")
                   .str.replace("million","")
                  )
nauru = nauru.drop(['Grants In Kind (USD)'], axis=1)
nauru['total'] = (nauru['Total']
                   .str.replace("$","")
                   .str.replace("million","")
                  )
nauru = nauru.drop(['Total'], axis=1)
nauru = nauru.fillna(0)
nauru
```

```python
nauru['country'] = 'nauru'
nauru
```

```python
nauru.to_csv('nauru-clean.csv')
```

```python
nauru_pivot = nauru.pivot(index='country', columns = 'Year', values = 'total')
nauru_pivot
```

```python
marshall['country'] = 'marshall'
```

```python
marshall_pivot = marshall.pivot(index='country', columns = 'Year', values = 'total')
marshall_pivot = marshall_pivot
marshall_pivot
```

```python
tuvalu = pd.read_csv('tuvalu-taiwan.csv')
tuvalu
```

```python
tuvalu['total'] = (tuvalu['USD']
                   .str.replace("$","")
                   .str.replace("million","")
                  )
tuvalu = tuvalu.drop(['USD'], axis=1)
tuvalu['country'] = 'tuvalu'
tuvalu = tuvalu.fillna(0)
tuvalu
```

```python
tuvalu
```

```python
tuvalu['country'] = 'tuvalu'
tuvalu = tuvalu.fillna(0)
tuvalu = tuvalu.drop([10])
tuvalu
```

```python
tuvalu.to_csv('tuvalu-clean.csv')
```

```python
tuvalu_pivot = tuvalu.pivot(index='country', columns = 'Year', values = 'total')
tuvalu_pivot
```

```python
combined = pd.concat([
    marshall_pivot,
    nauru_pivot,
    tuvalu_pivot
]).astype(float)
combined
# ignore_index=True
# combined.head()
```
